# 데이터 핸들링 - 판다스(Pandas) 기본 지식


판다스는 행과 열로 이뤄진 2차원 데이터를 효율적으로 가공 / 처리할 수 있는 기능을 제공하는 패키지입니다.
넘파이는 저수준 API가 대부분인 반면 판다스는 많은 부분이 넘파이 기반으로 작성됐으나 훨씬 유연하고 편리하게 데이터 핸들링이 가능합니다.
파이썬의 리스트, 컬렉션 넘파이 등의 내부 데이터뿐만 아니라 CSV 등의 파일을 쉽게 DataFrame으로 변경해줍니다.


**판다스의 핵심객체**
1. DataFrame : 여러 개의 행과 열로 이뤄진 2차원 데이터를 담는 데이터 구조체
2. Index : 개별 데이터를 고유하게 식별하는 Key값
3. Series : 칼럼이 하나뿐인 데이터 구조체 (DataFrame은 여러 개의 Series로 이뤄졌다고 할 수 있음)

```python
import pandas as pd
```

다양한 포맷으로 된 파일을 DataFrame으로 로딩할 수 있는 편리한 API 제공

1. read_csv() : CSV(칼럼을 ','로 구분한 파일 포맷)파일 포맷 변환을 위한 API
2. read_table() : 칼럼을 구분하는 필드 구분 문자(Delimeter)가 탭 문자인 경우
3. read_fxf() : Fixed Width, 즉 고정 길이 기반의 칼럼 포맷을 DataFrame으로 로딩하기 위한 API
    
만약 read_csv(sep = '\t')로 하면 read_table과 같은 기능.

### DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환


#### 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

DataFrame은 리스트와 넘파이 ndarray와 다르게 칼럼며을 가지고 있음.
일반적으로 DataFrame으로 변환 시 칼럼명을 지정해 줌(지정 하지 않으면 자동으로 할당)

**DataFrame**생성 인자
1. 리스트, 딕셔너리 또는 ndarray 입력
2. columns : 칼럼명 리스트 입력

DataFame은 기본저긍로 행과 열을 가지는 2차원 데이터 이므로 2차원 이하의 ㄷ3ㅔ이터들만 변환 가능.

```python
import numpy as np

# 1차원 형태 변환
col_name1 = ['col1']
list1 = [1, 2, 3]
array1 = np.array(list1)
print('array1 shape : ', array1.shape)
#리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns = col_name1)
print('1차원 리스트로 만든 DataFrame : \n', df_list1)
# 넘파이 ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns = col_name1)
print('\n1차원 ndarray로 만든 DataFrmae : \n', df_array1)
```

1차원 형태의 데이터를 기반으로 DataFrame 생성 시 칼럼명이 한 개만 필요하다는 사실 주의!

2차원 형태의 DataFrame의 경우 칼럼명은 열의 개수만큼 필요!

```python
# 2행 3열 형태 변환 -> 3개의 칼럼명 필요

col_name2 = ['col1', 'col2', 'col3']

# 2행 X 3열 형태의 리스트와 ndarray 생성 후 이를 DataFrame으로 변환
list2 = [[1, 2, 3],
         [11, 12, 13]]
array2 = np.array(list2)
print('array2 shape : ', array2.shape)
df_list2 = pd.DataFrame(list2, columns = col_name2)
print('2차원 리스트로 만든 DataFrame : \n', df_list2)
df_array2 = pd.DataFrame(array2, columns = col_name2)
print('\n2차원 ndarray로 만든 DataFrame : \n', df_array2)
```

딕셔너리의 경우 딕셔너리의 키(Key)가 칼럼명, 값(Value)가 칼럼 데이터로 변환.

```python
dict = {'col1' : [1, 11], 'col2' : [2, 22], 'col3' : [3, 33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame : \n', df_dict)
```

### DataFrame을 넘파이 ndarray,리스트, 딕셔너리로 변환하기

1. ndarray로 변환 : values
2. 리스트로 변환 : values.tolist() -> values로 얻은 ndarray를 리스트로 변환
3. 딕셔너리로 변환 : to_dict() *만약 인자로 'list' 입력 시 딕셔너리의 값이 리스트형으로 반환

```python
# DataFrame을 ndarray로 변환
array3 = df_dict.values
print('df_dict.values 타입 : ', type(array3), 'df_dict.values shape : ', array3.shape)
print(array3)

# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입 : ', type(list3))
print(list3)

# DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입 :', type(dict3))
print(dict3)
```

#### DataFrame ix[] 연산자와 loc[], iloc[]연산자

ix[행 위치, 열 위치] -> 행 위치 = index값, 열 위치 = 칼럼명(또는 칼럼의 위치 값)을 입력해 원하는 위치의 데이터 추출 가능

그러나 이 방식이 가독성이 떨어지고 코드 혼동으로 인해 현재는 deprecated되고 loc[]과 iloc[]연산자가 만들어짐.

#### iloc[] 연산자
- 위치 기반 인덱싱만 허용 -> 행과 열 값으로 integer 또는 integer 형의 슬라이싱, 팬시 리스트 값을 입력해줘야 함. 


```python
data = {'Name' : ['Chulmin', 'Eunkyung', 'Jinwoong', 'Soobeom'],
        'Year' : [2011, 2016, 2015, 2015],
        'Gender' : ['Male', 'Female', 'Male', 'Male']
       }
data_df = pd.DataFrame(data, index = ['one', 'two', 'three', 'four'])
data_df
```

```python
data_df.iloc[0, 0]

#다음 코드는 오류 발생

#data_df.iloc[0, 'Name']
#data_df.iloc['one', 0]
```

#### loc[] 연산자
- 명칭 기반 데이터 추출 -> 행 위치에는 DataFrame index값, 열 위치에는 칼럼 명

```python
data_df.loc['one', 'Name']

# 다음 코드는 오류 발생
#data_df.loc[0, 'Name']
```

loc[]에 슬라이싱 기호 ':'를 적용할 때 유의할 점!
- 일반적으로으로는 시작 값 ~ 종료값-1 까지를 의미하나 loc에서는 종료값까지 포함!
- 이는 명칭은 숫자형이 아닐 수도 있으므로 -1이 불가하기 때문!

```python
print('위치 기반 iloc slicing \n', data_df.iloc[0:1, 0], '\n')
print('명칭 기반 loc slicing \n', data_df.loc['one' : 'two', 'Name'])
```

### 결손 데이터(Missing Data) 처리하기

결손 데이터는 칼럼에 값이 없는, 즉 NULL인 경우 의미. 넘파이의 NaN으로 표시.

기본적으로 머신러닝 알고리즘은 이 NaN 값을 처리하지 않으므로 이 값을 다른 값으로 대체해야 함.
또한 NaN 값은 평균, 총합 등의 함수 연산 시 제외 됨.

- isna() : NaN 여부 확인
- fillna() : NaN 다른 값으로 대체

#### isna()로 결손 데이터 여부 확인

#### fillna()로 결손 데이터 대체하기

```python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
```

### apply lambda 식으로 데이터 가공
- apply 함수에 lambda 식을 결합해 DataFrame이나 Series의 레코드별로 데이터를 가공하는 기능 제공
- 판다스의 경우 칼럼에 일괄저그로 데이터 가공을 하는 것이 속도 면에서 더 빠르나 복잡한 데이터 가공이 필요해 apply lambda를 어쩔 수 없이 이용해야 함

```python
# lambda란? 간단 예제!

def get_square(a):
    return a**2

print('3의 제곱은 : ', get_square(3))

# 이러한 함수를 한 줄로 처리!
lambda_square = lambda x : x ** 2
print('3의 제곱은 : ', lambda_square(3))

# 여러 개의 값을 입력 인자로 사용시 map() 함수!
a = [1, 2, 3]
squares = map(lambda x : x ** 2, a)
list(squares)
```

```python
# 판다스에 lambda 적용해보기 

titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
```

lambda 식은 if else를 지원해줌!

단, 주의할 점! 
1. if 절의 경우 **if식보다 반환 값**을 먼저 기술해야함. lambda 식 ':' 기호의 오른 편에 반환 값이 있어야 하므로!!
2. if, else만 지원하고 else if는 지원하지 않음! 


* else if 이용하기 위해서는 else절을 ()로 내포해 다시 if else 적용해야 함.

```python
# 15세 이하는 Child, 15~60세는 Adult, 61 이상은 Elderly로 분류

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else ('Adult' if x <=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

else if가 많이 나와야 하는 경우 else를 계속 내포해서 쓰기에는 부담스러움.

이 경우에는 아예 별도의 함수를 만드는 것을 추천!!

```python
def get_category(age) :
    cat = ''
    if age <= 5 : cat = 'Baby'
    elif age <= 12 : cat = 'Child'
    elif age <= 18 : cat = 'Teenager'
    elif age <= 25 : cat = 'Student'
    elif age <= 35 : cat = 'Young Adult'
    elif age <= 60 : cat = 'Adult'
    else : cat = 'Elderly'
        
    return cat

# lambda 식에 위에서 생성한 get_category()함수를 반환값으로 지정
# get_category(x)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
